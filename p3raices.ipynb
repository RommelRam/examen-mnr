```python
#graficar la siguiente grafica
def f(x):
    return  x**5 + 5.25*x**4 + 4.125 * x**3 - 9.125 * x**2 - 14.625 * x + 5.625

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 1.5, 100)
y = f(x)
plt.plot(x, y, label='f(x)')
plt.axhline(0, color='red', linestyle='--', label='y=0')
plt.title('Gráfica de f(x)')  
```

```python
import numpy as np
from scipy.optimize import newton

def f(x):
    return  x**5 + 5.25*x**4 + 4.125 * x**3 - 9.125 * x**2 - 14.625 * x + 5.625
def df(x):
    return 5*x**4 + 21*x**3 + 12.375 * x**2 - 18.25 * x - 14.625
root = newton(f, x0=-4, fprime=df)
print(f"Raíz aproximada: {root}")
```
